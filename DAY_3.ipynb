## Iris Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

##Univariate Analysis

```python
sns.histplot(df['sepal_length'],bins=10,kde=True,color='skyblue')
plt.title("Sepal Length Distribution")
plt.show()
```

```python
sns.countplot(x='species',data=df,color='skyblue',palette="Set2",edgecolor='black')
plt.title("Species Distribution")
plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
sns.pairplot(df, hue='species', diag_kind="kde")
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Sepal Length by Species")
plt.show()
```

```python
sns.violinplot(x='species', y='petal_length', data=df, split=True)
plt.title("Petal Length by Species (Violin)")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
