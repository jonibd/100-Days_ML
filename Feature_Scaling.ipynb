```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
data={
    'Age':[25, 30, 35, 40, 45, 50, 55, 60],
     'Years_Experience': [2, 5, 8, 12, 15, 18, 21, 25],
    'Salary': [50000, 60000, 75000, 90000, 110000, 130000, 150000, 180000],
    'Purchased': [0, 1, 0, 1, 0, 1, 0, 1]
}
```

```python
df=pd.DataFrame(data)
df
```

**divide feather and target**

```python
x=df[['Age', 'Years_Experience', 'Salary']]
y=df['Purchased']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

#**: StandardScaler**

```python
snds=StandardScaler()
x_train_S_S=snds.fit_transform(x_train)
x_test_s_s=snds.transform(x_test)
```

```python
x_train_S_S
```

```python
x_test_s_s
```

```python
X_train_scaled_std = pd.DataFrame(x_train_S_S, columns=x_train.columns)
X_test_scaled_std = pd.DataFrame(x_test_s_s, columns=x_test.columns)
```

```python
X_train_scaled_std
```

```python
X_test_scaled_std
```
