```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
data = {
    'Age': [25, 30, np.nan, 35, 40, np.nan, 45],
    'Salary': [50000, 60000, 70000, np.nan, 90000, 110000, 120000],
    'Country': ['India', 'USA', 'UK', np.nan, 'USA', 'India', 'UK'],
    'Purchased': ['No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No']
}
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
x=df[['Age','Salary','Country']]
y=df['Purchased']
```

```python
X_train, X_test, y_train, y_test = train_test_split(  x,y, random_state=42, test_size=0.2)
```

```python
imputer_numeric=SimpleImputer(strategy='median')
```

```python
imputer_categorical=SimpleImputer(strategy='most_frequent')
```

```python
numeric_cols = ['Age', 'Salary']
categorical_cols = ['Country']
```

```python
X_train[numeric_cols]=imputer_numeric.fit_transform(X_train[numeric_cols])
X_test[numeric_cols]=imputer_numeric.transform(X_test[numeric_cols])
```

```python
X_train[categorical_cols] = imputer_categorical.fit_transform(X_train[categorical_cols])
X_test[categorical_cols] = imputer_categorical.transform(X_test[categorical_cols])
```
