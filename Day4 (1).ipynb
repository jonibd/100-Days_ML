```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Sample Data
data = pd.DataFrame({
    'Age': [22, 35, 58, 45, 30],
    'Fare': [7.25, 71.83, 8.05, 512.33, 13.0]
})

# Scaling  
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
scaled_df = pd.DataFrame(scaled_data, columns=['Age', 'Fare'])

# Plot
plt.figure(figsize=(12,5))

# Original Data
plt.subplot(1,2,1)
plt.scatter(data['Age'], data['Fare'], color='blue')
plt.title("Original Data")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.grid(True)

# Scaled Data
plt.subplot(1,2,2)
plt.scatter(scaled_df['Age'], scaled_df['Fare'], color='red')
plt.title("Standardized Data")
plt.xlabel("Age (scaled)")
plt.ylabel("Fare (scaled)")
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(12,5))

# Original
plt.subplot(1,2,1)
data['Fare'].hist(color='blue', bins=5)
plt.title("Original Fare Distribution")

# Scaled
plt.subplot(1,2,2)
scaled_df['Fare'].hist(color='red', bins=5)
plt.title("Scaled Fare Distribution")

plt.show()
```

# normalization**

# Min-Max normalization (most common):
	


```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python


# Example data
data = pd.DataFrame({
    'Age': [22, 35, 58, 45, 30],
    'Fare': [7.25, 71.83, 8.05, 512.33, 13.0]
})
data
```

```python
sc=MinMaxScaler()
normalized_data=sc.fit_transform(data)
```

```python
normalized_df = pd.DataFrame(normalized_data, columns=['Age', 'Fare'])
```

```python
normalized_df
```

```python
normalized_df.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(data['Age'], bins=5, color='lightblue')
plt.title("Before Normalization - Age")

```

```python
plt.subplot(1,2,2)
plt.hist(normalized_df['Age'], bins=5, color='lightgreen')
plt.title("After Normalization - Age")
plt.show()
```

#Robust Scaler**

```python
from sklearn.preprocessing import RobustScaler
import pandas as pd
```

```python

data = pd.DataFrame({
    'Age': [22, 25, 30, 28, 100],  # Notice 100 is an outlier
    'Salary': [20000, 25000, 30000, 28000, 100000]  # 100000 outlier
})
data
```

```python
scaler= RobustScaler()
scaled_data= scaler.fit(data)
robust_data=scaler.transform(data)
```

```python
robust_data
```

```python
robust_df=pd.DataFrame(robust_data, columns=['Age','Salary'])
```

```python
robust_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.boxplot(data['Age'])
plt.title("Before Scaling (Age)")

plt.subplot(1,2,2)
plt.boxplot(robust_df['Age'])
plt.title("After Robust Scaling (Age)")
plt.show()
```
