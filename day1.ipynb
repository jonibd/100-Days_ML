# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]   
y = [2, 4, 6, 8, 10]

plt.plot(x,y, color="green",linestyle="--",marker="o")
plt.title("Simple Line Graph")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]

plt.plot(x, [i for i in x], label="y=x")        # y=x
plt.plot(x, [i**2 for i in x], label="y=x^2")  # y=x^2
plt.plot(x, [i**3 for i in x], label="y=x^3")  # y=x^3

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Multiple Lines Example")
plt.legend()   # লাইনগুলোর নাম দেখাবে
plt.show()
```

```python
x = ["A", "B", "C", "D"]
y = [5, 7, 3, 8]

plt.bar(x, y, color="green")
plt.title("Bar Chart Example")
plt.show()
```

```python
sizes = [30, 25, 20, 25]
labels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(sizes,labels=labels,autopct="%1.1f%%")
plt.title("Pie Chart Example")
plt.show()
```

```python
x = [5, 7, 8, 7, 6, 9, 5, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87]

plt.scatter(x, y, color="purple")
plt.title("Scatter Plot Example")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(8,5))   # গ্রাফের সাইজ পরিবর্তন
plt.plot(x, y, marker="o")
plt.grid(True)              # grid line যোগ করা
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Dataset load করা
df = sns.load_dataset("tips")
print(df.head())
```

```python
df.shape
```

```python
sns.scatterplot(x="total_bill",y="tip",color='red',marker='*',data=df)
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

```python
sns.histplot(df["total_bill"], bins=20,kde=True )
plt.title("Histogram of Total Bill")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Box Plot: Day vs Total Bill")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=df)
plt.title("Average Bill per Day")
plt.show()
```

```python
# correlation বের করা
corr = df.corr(numeric_only=True)

# heatmap আঁকা
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[["total_bill", "tip", "size"]])
plt.show()
```
